```python
# Importação da biblioteca requests para realizar requisições HTTP
import requests

# Importação da biblioteca pandas para manipulação de dados
import pandas as pd

import time

# Definição da chave da API Alpha Vantage
API_KEY = 'JP4P5N24KKIJVKVP'

# Lista de empresas (símbolos das ações)
EMPRESAS = ['NFLX', 'AAP', 'GOOGL', 'AMZN', 'MSFT']

# Lista para armazenar os DataFrames de cada empresa
lista_emp = []

# Loop para percorrer cada símbolo
for simbolo in EMPRESAS:
    print(f'Extraindo dados de {simbolo}')

    # Construção da URL com o símbolo atual
    # function=GLOBAL_QUOTE retorna apenas o valor mais recente
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={simbolo}&apikey={API_KEY}'

    # Requisição para a API
    r = requests.get(url)
    base = r.json()

    # Verificação da existência da chave 'Global Quote'
    if 'Global Quote' in base:
        # Conversão para DataFrame (estrutura de linha única)
        base_alpha = pd.DataFrame([base['Global Quote']])

        # Adição do DataFrame à lista
        lista_emp.append(base_alpha)
    else:
        print(f'Erro ou limite atingido para {simbolo}:', base)

    time.sleep(3)  # evita limite da API

# Concatenação de todos os DataFrames em um único
base_alpha_empresas = pd.concat(lista_emp, ignore_index=True)

# Ajuste dos nomes das colunas:
# remove numeração inicial (ex: '01. symbol' → 'symbol')
base_alpha_empresas.columns = [col.split('. ')[1] for col in base_alpha_empresas.columns]

# Definição das colunas numéricas
cols_numerics = ['open', 'high', 'low', 'price', 'volume', 'previous close', 'change']

# Conversão das colunas numéricas para float
base_alpha_empresas[cols_numerics] = base_alpha_empresas[cols_numerics].astype(float)
```
